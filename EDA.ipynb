```python
print(df.isna())
df["Item_Weight"].fillna(df["Item_Weight"].mean(), inplace=True)
df.isna().sum()
```

```python
df["Outlet_Size"].fillna(0, inplace=True)
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
frequency = df["Outlet_Size"].value_counts()
percentage = df["Outlet_Size"].value_counts(normalize=True) * 100
mode = df["Outlet_Size"].mode()[0]
print(mode+'\n')
print(percentage)
print(frequency)
```

```python
frequency.plot(kind="bar", color="red")
plt.title("Outlet Size Frequency")
plt.xlabel("Outlet_size")
plt.ylabel("frequency")
plt.show()
sns.countplot(x=df["Outlet_Size"],color="red")
```
